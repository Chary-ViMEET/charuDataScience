```python
#implementation of Gradient Descent in Linear Regression
import matplotlib.pyplot as plt
import pandas as pd
#data = [[1,2,3,4,12,21,7,45,19,91],[5,6,7,8,91,5,77,70,17,23]]
data = pd.read_csv("girls_train.csv")
b0 = 0
b1 = 0
dt0 = []
dt1 = []
print(data.iloc[6,0])
for i in range(1,1500):
    for j in range(0,50):
        t1 = -data.iloc[j,0]*(data.iloc[j,1]-(b0*data.iloc[j,0] + b1))
        t2 = -(data.iloc[j,1] -(b0 * data.iloc[j,0] + b1))
    b0 = b0 - 0.005 * (2 / 60.0) * t1
    b1 = b1 - 0.005 * (2 / 60.0) * t2
    dt0.append(b0)
    dt1.append(b1)
    #print("B0:",b0,"B1:",b1)
    #plotting GD
print("printing Beta0")
print(dt0)
print("printing Beta1")
print(dt1)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(dt0,dt1,"-" ) 
#ax.plot(dt0,dt0)
ax.plot(data)
plt.show()
```
